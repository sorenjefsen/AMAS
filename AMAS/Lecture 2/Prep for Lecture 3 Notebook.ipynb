```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('ggplot')
```

```python
class Sampler():
    def __init__(self):
        None

    def sample_uniform(self, interval = [0, 9.7], n_samples = 100):
        self.uniform_samples = np.random.uniform(interval[0], interval[1], n_samples)
        return self.uniform_samples
    
    def sample_gaussian(self, mu = 0, sigma = np.sqrt(1.2), n_samples = 103, method = 'numpy'):
        if method == 'numpy':
            self.gaussian_samples = np.random.normal(mu, sigma, n_samples)
        
    
    def plot_samples(self):
        if not hasattr(self, 'uniform_samples'):
            self.sample_uniform()
        if not hasattr(self, 'gaussian_samples'):
            self.sample_gaussian()

        fig, ax = plt.subplots(1,2, figsize = (10,4))
        ax[0].hist(self.uniform_samples, bins = 10)
        ax[0].set_title("Uniform samples")
        ax[1].hist(self.gaussian_samples, bins = 10)
        ax[1].set_title("Gaussian samples")

    def save_txt(self, samples, filename):
        np.savetxt(filename, samples)


sampler = Sampler()
#sampler.sample_uniform()
#sampler.sample_gaussian()
sampler.plot_samples()
```
